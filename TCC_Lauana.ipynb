```python
def main(path):
  #load df
  df = pd.read_csv(path)
  df_count = pd.DataFrame() #create void df

  for i in df.columns: #preprocessed df, count values of df and append on df_count
      value_counts_normalized = df[i].value_counts(normalize=True)
      df_count = df_count.append(value_counts_normalized)

  df_count.fillna(0,inplace=True)# removing NaB values
  df_count=df_count[sorted(df_count.columns)] #sorting df_count columns


  #select the questions of df
  perguntas_= df_count.index.tolist()

  #make the graph plot
  figura = plot_Bar(df_count,perguntas_)

  return df_count,perguntas_,figura
```

```python
def plot_Bar(data,statements):
  categories = [1, 2, 3, 4, 5, 6, 7]
  data_x = [[round(val * 100, 2) for val in row] for row in data.values]

  # Ordenar os dados e os rótulos
  sorted_data, sorted_statements = zip(*sorted(zip(data_x, statements), key=lambda x: x[1]))
  # Criar o gráfico de barras
  fig = go.Figure()

  for i, category in enumerate(categories):
      fig.add_trace(go.Bar(
          y=sorted_statements,
          x=[datum[i] for datum in sorted_data],
          name=str(category),
          orientation='h',
          text=[f"{datum[i]:.2f}%" if datum[i] != 0 else "" for datum in sorted_data],
          textposition='auto'
      ))

  fig.update_layout(
        barmode='stack',
        xaxis={'categoryorder': 'total descending'},
        height=800,
        title={'text': 'Título do Gráfico', 'font': {'size': 24}},  # Ajuste o tamanho da fonte do título
        xaxis_title={'text': 'Rótulo do Eixo X', 'font': {'size': 18}},  # Ajuste o tamanho da fonte do rótulo do eixo x
        yaxis_title={'text': 'Rótulo do Eixo Y', 'font': {'size': 18}},  # Ajuste o tamanho da fonte do rótulo do eixo y
        legend={'font': {'size': 14}},  # Ajuste o tamanho da fonte da legenda
    )

  return fig
```

```python
df = pd.read_csv("/content/TCC 2 Coleta de feedback sobre o uso dos Badges para em atividades para gamificação no ensino de Qualidade de Software (respostas) - Elementos da gamificação.csv")
```

```python
df
```

```python
df_count = pd.DataFrame() #create void df

for i in df.columns: #preprocessed df, count values of df and append on df_count
    value_counts_normalized = df[i].value_counts(normalize=True)
    df_count = df_count.append(value_counts_normalized)

df_count.fillna(0,inplace=True)# removing NaB values
df_count=df_count[sorted(df_count.columns)]
```

```python
df_count
```

```python
# Reordenar as colunas
df_count=df_count[['Insatisfeito', 'Neutro', 'Satisfeito', 'Muito satisfeito']]
```

```python
statements= df_count.index.tolist()
```

1 (não é verdadeira), 2 (Pouco verdadeira) , 3 (Parcialmente não é  verdadeira), 4 (um pouco verdadeira), 5 (Parcialmente verdadeira), 6 (Bastante Verdadeira), 7 (Muito verdadeira)

```python
df_count.columns
```

```python
categories = ['Insatisfeito', 'Neutro', 'Satisfeito', 'Muito satisfeito']
data_x = [[round(val * 100, 2) for val in row] for row in df_count.values]

# Ordenar os dados e os rótulos
sorted_data, sorted_statements = zip(*sorted(zip(data_x, statements), key=lambda x: x[1]))
# Criar o gráfico de barras
fig = go.Figure()

#colors = ['#FF7F7F', '#9999FF', '#00FF00', '#7F7FFF']
#colors = ['#FF0000', '#008000', '#0000FF', '#7F7FFF']

colors = ['#EF553B', '#636EFA', '#00CC96', '#AB63FA'] # Cores invertidas para Insatisfeito, Neutro e Satisfeito

for i, category in enumerate(categories):
    color = colors[i] if category in ['Insatisfeito', 'Neutro', 'Satisfeito'] else '#AB63FA'
    fig.add_trace(go.Bar(
        y=sorted_statements,
        x=[datum[i] for datum in sorted_data],
        name=str(category),
        orientation='h',
        text=[f"{datum[i]:.2f}%" if datum[i] != 0 else "" for datum in sorted_data],
        textposition='auto',
        marker_color=color  # Adicionado esta linha para definir a cor com base na categoria
    ))

fig.update_layout(barmode='stack', xaxis={'categoryorder':'total descending'},height=800)
fig.show()
```
